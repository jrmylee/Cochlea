```python
import librosa
import numpy as np
from IPython.display import Audio
```

```python
x, sr = librosa.load("samples/widmung.mp3")
```

```python
hop_length = 2048
n_fft = 2048
```

```python
clip = x[20 * sr : 22 * sr]
```

```python
stft = librosa.stft(clip)
```

```python
db = (-1 * librosa.amplitude_to_db(stft, ref=np.max)) / 120
```

```python
amp = librosa.db_to_amplitude(db)
```

```python
y = librosa.griffinlim(amp)
```

```python
Audio(y, rate=22050)
```
